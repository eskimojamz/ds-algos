```python
class Node:
    def __init__(self, data=None, prev=None, next=None):
        self.data = data
        self.prev = prev
        self.next = next
    
class DoublyLinkedList:
    def __init__(self):
        self.head = None
    
    def append(self, data):
        if self.head is None:
            self.head = Node(data)
            return
        # There is already head node
        cur = self.head
        new_node = Node(data)
        # iterate until cur.next is None (end of list)
        while cur.next:
            cur = cur.next
        cur.next = new_node
        new_node.prev = cur

    def prepend(self, data):
        if self.head is None:
            self.head = Node(data)
            return
        #There is already head node
        cur = self.head
        new_node = Node(data)
        # Set new_node.next, cur.prev, and self.head
        cur.prev = new_node
        new_node.next = cur
        self.head = new_node

    def add_after(self, key, data):
        cur = self.head
        while cur:
            if cur.next is None and cur.data == key:
                self.append(data)
                return
            elif cur.data == key:
                nxt = cur.next
                new_node = Node(data)
                cur.next = new_node
                new_node.prev = cur
                new_node.next = nxt
                nxt.prev = new_node
            cur = cur.next
        
    def add_before(self, key, data):
        cur = self.head
        while cur:
            if cur.prev is None and cur.data == key:
                self.prepend(data)
                return
            elif cur.data == key:
                prev = cur.prev
                new_node = Node(data)
                prev.next = new_node
                new_node.prev = prev
                new_node.next = cur
                cur.prev = new_node
            cur = cur.next

    def print_forward(self):
        if self.head is None:
            print('Empty List')
            return 
        cur = self.head
        llstr = ''
        while cur:
            llstr += str(cur.data) + ' ---> '
            cur = cur.next
        print(llstr)

    def get_last_node(self):
        cur = self.head
        while cur.next:
            cur = cur.next
        return cur

    def print_backward(self):
        if self.head is None:
            print('Empty list')
            return
        # Start from last node
        cur = self.get_last_node()
        llstr = ''
        # Iterate to end reverse direction by cur = cur.prev 
        while cur:
            llstr += str(cur.data) + ' ---> '
            cur = cur.prev
        print(llstr)
        

dll = DoublyLinkedList()
dll.append('a')
dll.append('b')
dll.append('c')
dll.prepend('aa')
dll.add_after('b', 'bb')
dll.add_before('c', 'bbb')
dll.print_forward()
```
