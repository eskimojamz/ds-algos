```python
class HashTable:  
    def __init__(self):
        self.MAX = 10
        self.arr = [[] for i in range(self.MAX)]
        
    def get_hash(self, key):
        hash = 0
        for char in key:
            hash += ord(char)
        return hash % self.MAX
    
    def __getitem__(self, key):
        h = self.get_hash(key)
        for element in self.arr[h]:
            if len(element)==2 and element[0] == key:
                print(element[0], ':', element[1])
    
    def __setitem__(self, key, val):
        h = self.get_hash(key)
        found = False
        # if found, it already exists
        # index and element in self.arr[h]
        for idx, element in enumerate(self.arr[h]):
            # if key of element matches, changing the existing key
            if len(element)==2 and element[0] == key:
                #for that index of arr[h](key, val) update it
                self.arr[h][idx] = (key, val)
                found = True
                break
        # it doesn't exist
        if not found:
            self.arr[h].append((key, val))  
```

```python
t = HashTable()
t["march 6"] = 310
t["march 7"] = 420
t["march 8"] = 67
t["march 17"] = 63457
```
