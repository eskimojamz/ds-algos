```python
class HashTable:  
    def __init__(self):
        self.MAX = 10
        self.arr = [None for i in range(self.MAX)]
        
    def get_hash(self, key):
        hash = 0
        for char in key:
            hash += ord(char)
        return hash % self.MAX
    
    def __getitem__(self, key):
        h = self.get_hash(key)
        for element in self.arr[h]:
            return element
    
    def __setitem__(self, key, val):
        h = self.get_hash(key)
        # hash slot is empty
        if self.arr[h] is None:
            self.arr[h] = (key, val)
        else:
            new_h = self.find_slot(key, h)
            self.arr[new_h] = (key, val)
        print(self.arr)
    
    def __delitem__(self, key):
        for i, element in enumerate(self.arr):
            if element is not None and element[0] == key:
                print("Deleted:", element)
                self.arr[i] = None
                return
        raise Exception("Key doesn't exist")

    def get_probe_range(self, h):
        return [*range(h, len(self.arr))] + [*range(0, h)]
    
    def find_slot(self, key, h):
        probe_range = self.get_probe_range(h)
        for p in probe_range:
            # slot is empty, use to add key
            if self.arr[p] is None:
                return p
            # slot contains key, use slot to set new key
            if self.arr[p][0] == key:
                return p
            # if p is not empty or matches key, keep looping to find empty slot 
        raise Exception("Hashtable full")
```

```python
t = HashTable()
t["march 6"] = 310
t["march 17"] = 459
del t["march 6"]
print(t.arr)
```
