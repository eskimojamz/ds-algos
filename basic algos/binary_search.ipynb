```python
def binary_search(list, target):
    first = 0
    last = len(list) - 1

    while first <= last:
        midpoint = (first + last) // 2

        if list[midpoint] == target:
            return midpoint
        elif list[midpoint] < target:
            first = midpoint + 1
        else:
            last = midpoint - 1

    return -1
```

```python
list = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print('At Index: ', binary_search(list, 7))
```

```python
def binary_search_recursive(list, target, first, last):
    midpoint = (first + last) // 2

    if list[midpoint] == target:
        return midpoint
    elif list[midpoint] < target:
        first = midpoint + 1
    else:
        last = midpoint - 1
    
    return binary_search_recursive(list, target, first, last)
```

```python
list = [1, 2, 3, 4, 5, 6, 7, 8, 9]
binary_search_recursive(list, 7, 0, len(list) - 1)
```
